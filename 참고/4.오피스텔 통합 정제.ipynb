```python
import os
import pandas as pd

# 📌 설정부
base_path = r"C:\startcoding\GUKTO\오피스텔_매매"
output_folder = os.path.join(base_path, "오피스텔_매매_통합")
os.makedirs(output_folder, exist_ok=True)
output_path = os.path.join(output_folder, "오피스텔_전국통합.xlsx")

df_list = []

# 📌 폴더 순회
for region in os.listdir(base_path):
    region_path = os.path.join(base_path, region)
    
    if not os.path.isdir(region_path) or "통합" in region:
        continue
    
    print(f"📂 작업 중: {region}")
    
    for file in os.listdir(region_path):
        if file.endswith(".xlsx") and not file.startswith("~$"):
            file_path = os.path.join(region_path, file)
            
            try:
                # 📌 핵심 수정: 13행까지 무시(skiprows=13)하고, 컬럼명 없음(header=None) 설정
                # 이렇게 하면 14행 데이터부터 순수하게 값으로 들어와.
                df = pd.read_excel(file_path, skiprows=13, header=None, engine='openpyxl')
                
                # 📌 R열(18번째 열)까지만 슬라이싱
                df = df.iloc[:, :18]
                
                # 빈 행 제거 (데이터가 아예 없는 경우 방지)
                df = df.dropna(how='all')
                
                if not df.empty:
                    # 나중에 지역 구분을 위해 폴더명만 슬쩍 추가해둘게
                    df["지역명_폴더"] = region 
                    df_list.append(df)
                
            except Exception as e:
                print(f"   ㄴ ❌ 오류 ({file}): {e}")

# 📌 데이터 통합 및 저장
if df_list:
    merged_df = pd.concat(df_list, ignore_index=True)
    
    # 중복 제거 (전체 행이 똑같은 경우만)
    merged_df.drop_duplicates(inplace=True)
    
    # 엑셀 저장 (컬럼명 없이 저장하려면 header=False)
    # 하지만 나중에 정제할 때 편하려면 기본 숫자 헤더라도 있는 게 나을 수 있어.
    merged_df.to_excel(output_path, index=False)
    
    print(f"\n✨ 통합 완료! 총 {len(merged_df)}행의 데이터를 모았어.")
    print(f"📍 파일 위치: {output_path}")
else:
    print("\n⚠️ 통합할 데이터가 없어.")
```

```python
import pandas as pd
import numpy as np
import os

# 📌 경로 설정
input_folder = r"C:\startcoding\GUKTO\오피스텔_매매\오피스텔_매매_통합"
output_folder = r"C:\startcoding\GUKTO\오피스텔_매매\오피스텔_매매_정제"

os.makedirs(output_folder, exist_ok=True)

# 통합된 파일 리스트 가져오기
files = [f for f in os.listdir(input_folder) if f.endswith(".xlsx") and "통합" in f]

for file in files:
    print(f"🔄 오피스텔 정제 시작: {file}")
    path = os.path.join(input_folder, file)
    
    # 통합 시 제목줄을 버렸으므로 header=None으로 읽음
    df = pd.read_excel(path, header=None, engine='openpyxl')

    df_clean = pd.DataFrame()

    # 1. 컬럼 매핑 (오피스텔 구조 반영: 아파트의 '동' 컬럼이 없어 인덱스 조정됨)
    df_clean['시군구'] = df.iloc[:, 1]
    df_clean['번지'] = df.iloc[:, 2]
    df_clean['본번'] = df.iloc[:, 3]
    df_clean['부번'] = df.iloc[:, 4]
    df_clean['단지명'] = df.iloc[:, 5]
    df_clean['전용면적'] = df.iloc[:, 6]
    df_clean['계약년월'] = df.iloc[:, 7]
    df_clean['계약일'] = df.iloc[:, 8]
    df_clean['거래금액'] = df.iloc[:, 9]
    df_clean['층'] = df.iloc[:, 10]        # 아파트 11번에서 10번으로 변경
    df_clean['매수자'] = df.iloc[:, 11]      # 아파트 12번에서 11번으로 변경
    df_clean['매도자'] = df.iloc[:, 12]      # 아파트 13번에서 12번으로 변경
    df_clean['건축년도'] = df.iloc[:, 13]    # 아파트 14번에서 13번으로 변경
    df_clean['도로명'] = df.iloc[:, 14]      # 아파트 15번에서 14번으로 변경

    # 2. 해제거래 제거 (해제사유발생일: 인덱스 15번)
    if df.shape[1] > 15:
        cancel_val = df.iloc[:, 15].astype(str).str.strip()
        # 날짜 형식(8자리 숫자 등)이 있으면 제거
        df_clean = df_clean[~cancel_val.str.match(r'^\d{4,8}$')]

    # 3. 데이터 타입 변환 (거래금액 콤마 제거 및 숫자화)
    df_clean['거래금액'] = df_clean['거래금액'].astype(str).str.replace(',', '')
    num_cols = ['전용면적', '계약년월', '계약일', '거래금액', '층', '건축년도']
    for col in num_cols:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

    # 4. 날짜 및 연식 계산
    # 계약일이 NaN인 경우를 대비해 0으로 채우고 정수화
    day_str = df_clean['계약일'].fillna(1).astype(int).astype(str).str.zfill(2)
    temp_dt = pd.to_datetime(
        df_clean['계약년월'].astype(str) + day_str, 
        format='%Y%m%d', errors='coerce'
    )
    
    # 계약일자 (YYMMDD) 및 연식 계산
    df_clean['계약일자'] = temp_dt.dt.strftime('%y%m%d')
    df_clean['연식'] = temp_dt.dt.year - df_clean['건축년도']

    # 5. 주소 분해 (최대 5개)
    addr_split = df_clean['시군구'].astype(str).str.split(' ', expand=True)
    for i in range(5):
        df_clean[f'주소{i+1}'] = addr_split[i] if i < addr_split.shape[1] else ""

    # 6. 파생 변수 생성
    # 면적규모 (30단위), 연식규모 (10단위)
    df_clean['면적규모'] = (df_clean['전용면적'] / 30).round() * 30
    df_clean['연식규모'] = (df_clean['연식'] / 10).round() * 10
    
    # 단가 계산 (정수형, 0으로 나눌 경우 대비)
    df_clean['단가'] = (df_clean['거래금액'] / df_clean['전용면적']).replace([np.inf, -np.inf], np.nan).fillna(0).astype(int)

    # 7. 순서번호 및 최종 정리
    df_clean = df_clean.reset_index(drop=True)
    df_clean['순서번호'] = df_clean.index + 1

    # 최종 컬럼 순서 설정
    final_cols = [
        '순서번호', '주소1', '주소2', '주소3', '주소4', '주소5',
        '번지', '도로명', '단지명', '매수자', '매도자',
        '계약일자', '거래금액', '면적규모', '연식규모',
        '전용면적', '연식', '층', '단가'
    ]

    # 존재하는 컬럼만 선택 (에러 방지)
    df_final = df_clean[[c for c in final_cols if c in df_clean.columns]]

    # 8. 저장
    save_filename = file.replace("전국통합", "전국정제").replace(".xlsx", "_정제.xlsx")
    save_path = os.path.join(output_folder, save_filename)
    df_final.to_excel(save_path, index=False)
    print(f"✅ 정제 완료: {save_path}")

print("\n✨ 오피스텔 모든 파일 정제 완료!")
```
