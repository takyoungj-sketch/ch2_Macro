```python
import os
import glob
import pandas as pd

# 기준 폴더
base_path = r'C:\startcoding\GUKTO\공장창고_매매'

# 통합 폴더 생성
merge_folder = os.path.join(base_path, '공장창고_매매_통합')
os.makedirs(merge_folder, exist_ok=True)

# 시도 폴더 목록
folders = [
    f for f in os.listdir(base_path)
    if os.path.isdir(os.path.join(base_path, f))
    and f != '공장창고_매매_통합'
]

merged_df = pd.DataFrame()

for region in folders:

    region_path = os.path.join(base_path, region)

    excel_files = glob.glob(os.path.join(region_path, '*.xlsx'))

    for file in excel_files:

        try:
            df = pd.read_excel(
                file,
                skiprows=13,   # 유효행 시작 (14행)
                header=None
            ).iloc[:, :21]    # 21열까지

            merged_df = pd.concat([merged_df, df], ignore_index=True)

            print(f"추가 완료: {region} - {os.path.basename(file)}")

        except Exception as e:
            print(f"오류: {file} / {e}")

# 저장
output_path = os.path.join(
    merge_folder,
    '공장창고_매매_전국통합.xlsx'
)

merged_df.to_excel(output_path, index=False, header=False)

print("\n전국 통합 완료")
print(output_path)
```

```python
import pandas as pd
import os

base_path = r"C:\startcoding\GUKTO\공장창고_매매"

input_file = os.path.join(
    base_path,
    "공장창고_매매_통합",
    "공장창고_매매_전국통합.xlsx"
)

output_folder = os.path.join(
    base_path,
    "공장창고_매매_정제"
)

os.makedirs(output_folder, exist_ok=True)

output_file = os.path.join(
    output_folder,
    "공장창고_매매_정제.xlsx"
)

# 엑셀 읽기
df = pd.read_excel(input_file, header=None)

print("원본 행수:", len(df))

# 필요한 열만 선택
df = pd.DataFrame({

    "시군구": df.iloc[:,1],
    "유형": df.iloc[:,2],
    "번지": df.iloc[:,3],
    "용도지역": df.iloc[:,5],
    "건축물용도": df.iloc[:,6],
    "도로조건": df.iloc[:,7],
    "연면적": df.iloc[:,8],
    "대지면적": df.iloc[:,9],
    "거래금액": df.iloc[:,10],
    "계약년월": df.iloc[:,14],
    "건축년도": df.iloc[:,17],
    "해제사유": df.iloc[:,18]

})

# 숫자 변환
df["거래금액"] = (
    df["거래금액"]
    .astype(str)
    .str.replace(",", "", regex=False)
)

df["거래금액"] = pd.to_numeric(df["거래금액"], errors="coerce")

df["연면적"] = pd.to_numeric(df["연면적"], errors="coerce")
df["대지면적"] = pd.to_numeric(df["대지면적"], errors="coerce")
df["건축년도"] = pd.to_numeric(df["건축년도"], errors="coerce")

# 해제거래 제거
df["해제사유"] = df["해제사유"].astype(str).str.strip()

df = df[
    (df["해제사유"] == "-") |
    (df["해제사유"] == "nan")
]


# 주소 분해
addr = df["시군구"].astype(str).str.split(" ", expand=True)

for i in range(5):
    if i in addr.columns:
        df[f"주{i+1}"] = addr[i]
    else:
        df[f"주{i+1}"] = ""

# 거래연도
df["거래연도"] = df["계약년월"].astype(str).str[2:4]


# -----------------------------
# 연식 계산
# -----------------------------

contract_year = df["계약년월"].astype(str).str[:4].astype(int)
df["연식"] = contract_year - df["건축년도"]

df.loc[df["연식"] < 0, "연식"] = None

# 연식구분
def age_scale(x):

    if pd.isna(x):
        return None

    if x <= 10:
        return round(x/5)*5
    else:
        return round(x/10)*10

df["연식구분"] = df["연식"].apply(age_scale)


# 규모 함수
def scale_round(x):

    if x <= 300:
        return round(x/50)*50
    elif x <= 1000:
        return round(x/100)*100
    elif x <= 5000:
        return round(x/1000)*1000
    else:
        return round(x/10000)*10000


df["건축규모"] = df["연면적"].apply(scale_round)
df["대지규모"] = df["대지면적"].apply(lambda x: scale_round(x) if pd.notna(x) else None)

# 도로 코드
def road_convert(x):

    if pd.isna(x):
        return ""

    x = str(x)

    if "4m미만" in x:
        return 2
    elif "8m미만" in x:
        return 6
    elif "25m미만" in x:
        return 23
    elif "25m이상" in x:
        return 27
    else:
        return ""


df["도로"] = df["도로조건"].apply(road_convert)

# 금액
df["금액"] = df["거래금액"]

# 용도지역 → 용적률 칼럼
# df["용적률"] = df["용도지역"]

# 최종 데이터
result = df[
[
"주1","주2","주3","주4","주5",
"번지",
"유형",
"거래연도",
"용도지역",
"건축물용도",
"건축규모",
"대지규모",
"연식구분",
"금액",
"연면적",
"대지면적",
"연식",
"도로"
]
]

# 저장
result.to_excel(output_file, index=False)


print("정제 완료")
print(output_file)
```
