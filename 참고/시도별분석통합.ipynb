```python
## 1. xlsx 통합 기능만 수행

import pandas as pd
import os
import glob

sido_name = '충청남도상업'
folder_path = fr'C:\startcoding\practice\{sido_name}'
tonghap_folder = os.path.join(folder_path, f'1.{sido_name}_변환 및 통합')

# 📂 통합 대상 폴더 내 .xlsx 파일 목록 수집 (통합 파일 제외)
xlsx_files = [
    f for f in glob.glob(os.path.join(folder_path, '*.xlsx'))
    if os.path.basename(f) != f'{sido_name}_통합.xlsx'
]

if not xlsx_files:
    print("⚠️ 통합 대상 .xlsx 파일이 없습니다.")
else:
    merged_df = pd.DataFrame()

    for file_path in xlsx_files:
        try:
            df = pd.read_excel(file_path, skiprows=13, header=None)
            df = df.iloc[:, :14]  # 앞 14개 열만 사용
            merged_df = pd.concat([merged_df, df], ignore_index=True)
            print(f"✅ 통합 중: {os.path.basename(file_path)}")
        except Exception as e:
            print(f"❌ 통합 실패: {os.path.basename(file_path)} - {e}")

    # 결과 저장
    output_path = os.path.join(tonghap_folder, f'{sido_name}_통합.xlsx')
    merged_df.to_excel(output_path, index=False, header=False, engine='openpyxl')
    print("✅ 모든 파일 통합 완료:", output_path)
```

```python
## 1. 시도별 파일 변환 후 통합: xls를 xlsx로 변환후 새로운 폴더에 저장하고 하나의 통합파일을 만든다. 

import pandas as pd
import os
import glob
import win32com.client as win32

sido_name = '충청남도상업'
folder_path = fr'C:\startcoding\practice\{sido_name}'
tonghap_folder = os.path.join(folder_path, f'1.{sido_name}_변환 및 통합')

# 📁 변환 및 통합 폴더 생성
os.makedirs(tonghap_folder, exist_ok=True)

# 🔄 .xls 파일을 Excel COM으로 .xlsx로 변환 후 저장
xls_files = glob.glob(os.path.join(folder_path, '*.xls'))

excel = win32.gencache.EnsureDispatch('Excel.Application')
excel.Visible = False

for xls_file in xls_files:
    try:
        wb = excel.Workbooks.Open(xls_file)
        file_name = os.path.basename(xls_file).replace('.xls', '.xlsx')
        xlsx_path = os.path.join(tonghap_folder, file_name)
        wb.SaveAs(xlsx_path, FileFormat=51)  # 51 = xlOpenXMLWorkbook (.xlsx)
        wb.Close(SaveChanges=False)
        print(f"✅ 빠른 변환 완료: {file_name}")
    except Exception as e:
        print(f"❌ 변환 실패: {os.path.basename(xls_file)} - {e}")

excel.Quit()

# 📂 새 폴더 내 .xlsx 파일들 통합 (통합.xlsx 제외)
xlsx_files = glob.glob(os.path.join(tonghap_folder, '*.xlsx'))
xlsx_files = [f for f in xlsx_files if not f.endswith('통합.xlsx')]

merged_df = pd.DataFrame()

for file_path in xlsx_files:
    try:
        df = pd.read_excel(file_path, skiprows=13, header=None)
        df = df.iloc[:, :14]
        merged_df = pd.concat([merged_df, df], ignore_index=True)
    except Exception as e:
        print(f"❌ 통합 실패: {os.path.basename(file_path)} - {e}")

# 결과 저장
output_path = os.path.join(tonghap_folder, f'{sido_name}_통합.xlsx')
merged_df.to_excel(output_path, index=False, header=False, engine='openpyxl')

print("✅ 모든 파일을 통합 완료:", output_path)
```

```python
## 2.시도별 전처리: 통합파일의 데이터 전처리

import pandas as pd
import os
sido_name = '대전광역시'

# 🔹 통합 엑셀 파일 경로
folder_path = fr'C:\startcoding\practice\{sido_name}'
folder_path2 = fr'C:\startcoding\practice\{sido_name}\1.{sido_name}_변환 및 통합'
file_path = os.path.join(folder_path2, f'{sido_name}_통합.xlsx')

# 🔹 데이터 불러오기
df = pd.read_excel(file_path, engine='openpyxl', header=None)

# 🔹 열 이름 지정
df.columns = ['순번', '시군구', '번지', '지목', '용도지역', '도로조건',
              '계약연월', '계약일', '계약면적', '거래금액(만원)', '지분구분',
              '해제사유발생일', '거래유형', '중개사소재지']

# 🔹 거래금액과 계약면적 클렌징 + 숫자 변환
for col in ['거래금액(만원)', '계약면적']:
    df[col] = df[col].astype(str).str.replace(',', '').str.strip()
    df[col] = pd.to_numeric(df[col], errors='coerce')

# 🔹 해제사유발생일 필터링 ('-'만 남기기)
df = df[df['해제사유발생일'] == '-']

# 🔹 중개사소재지 열 삭제
df = df.drop(columns=['중개사소재지'])

# 🔹 시군구 열 분리
split_columns = df['시군구'].str.split(' ', n=4, expand=True)
#split_columns.columns = ['주1', '주2', '주3', '주4', '주5']
split_columns.columns = [f'주{i+1}' for i in range(split_columns.shape[1])]

# 🔹 split한 열 추가
new_df = pd.concat([df, split_columns], axis=1)

# 🔹 계약연월 열 분리
new_df['연도'] = new_df['계약연월'].astype(str).str[:4]
new_df['월'] = new_df['계약연월'].astype(str).str[4:]

# 🔹 단가 열 추가 (만원/㎡, 소수점 1자리)
new_df['단가'] = (new_df['거래금액(만원)'] / new_df['계약면적']).round(1)

# 🔹 필요 없는 열 삭제
new_df = new_df.drop(columns=['계약연월', '해제사유발생일'])

# 🔹 열 순서 재배치
new_df = new_df[['순번', '시군구', '주1', '주2', '주3', '주4', '주5',
                 '번지', '지목', '용도지역', '도로조건',
                 '연도', '월',
                 '계약일', '계약면적', '거래금액(만원)', '단가',
                 '지분구분', '거래유형']]

# 🔹 용도지역 글자수 줄이기
# 예시 매핑 사전
zone_map = {
    '제1종전용주거지역': '1전',
    '제2종전용주거지역': '2전',
    '제1종일반주거지역': '1주',
    '제2종일반주거지역': '2주',
    '제3종일반주거지역': '3주',
    '준주거지역': '준주',
    '근린상업지역': '근상',
    '유통상업지역': '유상',
    '일반상업지역': '일상',
    '중심상업지역': '중상',
    '전용공업지역': '전공',
    '일반공업지역': '일공',
    '준공업지역': '준공',
    '자연녹지지역': '자녹',
    '생산녹지지역': '생녹',
    '보전녹지지역': '보녹',
    '계획관리지역': '계관',
    '보전관리지역': '보관',
    '생산관리지역': '생관',
    '개발제한구역': '개제',
    '농림지역': '농림',
    '자연환경보전지역': '자보'
}
# 매핑 적용
new_df['용도지역'] = new_df['용도지역'].map(zone_map).fillna('기타')

zone_map2 = {
    '과수원': '과',
    '창고용지': '창',
    '임야': '임',
    '공장용지': '장',
    '도로': '도',
    '잡종지': '잡',
    '구거': '구',
    '유지': '유',
    '학교용지': '학',
    '묘지': '묘',
    '철도용지': '철',
    '목장용지': '목',
    '제방': '제',
    '공원': '공',
    '양어장': '양',
    '하천': '천',
    '주차장': '차',
    '종교용지': '종',
    '수도용지': '수',
    '유원지': '원',
    '체육용지': '체',
    '주유소용지': '주',
    '사적지': '사',
    '염전': '염',     
}
# 매핑 적용

new_df['지목'] = new_df['지목'].apply(lambda x: zone_map2[x] if x in zone_map2 else x)

zone_map3 = {
    '-': '4미',
    '8m미만': '8미',
    '12m미만': '12미',
    '25m미만': '25미',
    '25m이상': '25이'
}
# 매핑 적용
new_df['도로조건'] = new_df['도로조건'].map(zone_map3).fillna('기타')

# 🔹 새로운 폴더에 저장
save_folder = os.path.join(folder_path, f'2.{sido_name}_전처리후')
os.makedirs(save_folder, exist_ok=True)  # 폴더 없으면 자동 생성

save_path = os.path.join(save_folder, f'{sido_name}_전처리후.xlsx')
new_df.to_excel(save_path, index=False, engine= 'openpyxl')

print('✅ 전처리 완료:', save_path)
```

```python
## 3.하위지역 필터

import pandas as pd
import os
import shutil
from openpyxl import load_workbook

# 🔹 분석 대상 행정구역 리스트
sido_name = '경기도'
target_areas = ['대장동']

# 🔹 파일 경로 설정
top_folder = r'C:\startcoding\practice'
folder_path = fr'C:\startcoding\practice\{sido_name}'
base_folder = fr'C:\startcoding\practice\{sido_name}\2.{sido_name}_전처리후'
file_path = os.path.join(base_folder, f'{sido_name}_전처리후.xlsx')
template_path = os.path.join(top_folder, '0.분석템플릿.xlsx')
filtered_folder = os.path.join(folder_path, f'3.{sido_name}_하위 필터')
os.makedirs(filtered_folder, exist_ok=True)

# 🔹 데이터 불러오기
df = pd.read_excel(file_path)

# 🔹 필터링 조건 적용
condition = df['시군구'].apply(lambda x: any(area in x for area in target_areas))
filtered_df = df[condition].copy()

# 🔹 결과 파일명 및 경로 설정
joined_name = '_'.join(target_areas).replace(' ', '')
save_path = os.path.join(filtered_folder, f'{joined_name}.xlsx')

# ✅ 템플릿 파일 복사 → 대상 파일로 (원본 보존)
shutil.copy(template_path, save_path)

# 🔹 복사본에 필터링 데이터 덮어쓰기 (Sheet1)
with pd.ExcelWriter(save_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    filtered_df.to_excel(writer, sheet_name='Sheet1', index=False)

# 🔹 결과 출력
print(f"✅ 포함된 행 개수: {len(filtered_df)}")
print(f"🔍 필터링된 구역: {', '.join(target_areas)}")
print(f"📁 저장 완료: {save_path}")
```

```python
## 4.분석

import pandas as pd
import numpy as np
from scipy.stats import norm
from openpyxl import load_workbook
import os

road_filter = '12미'           # 예: '대로'
cont_years = ''                # 예: '23, 24'

# 🔹 save_path는 앞 코드에서 이미 정의됨
# 🔹 템플릿 복사도 앞 코드에서 이미 완료됨

# 🔹 1. 데이터 로드
df = pd.read_excel(save_path, sheet_name='Sheet1')
df = df[['용도지역', '지목', '단가', '도로조건', '연도']].dropna()

# 🔹 2. 조건 필터링
if road_filter:
    df = df[df['도로조건'].str.contains(road_filter)]

if cont_years:
    year_list = [y.strip() for y in cont_years.split(',')]
    df = df[df['연도'].astype(str).str[-2:].isin(year_list)]

# 🔹 3. 통계 함수 정의
def stats(group):
    count = group.count()
    mean = group.mean()
    std = group.std()
    ci_low, ci_high = norm.interval(0.90, loc=mean, scale=std / np.sqrt(count)) if count > 1 else (np.nan, np.nan)
    q1 = group.quantile(0.25)
    median = group.median()
    q3 = group.quantile(0.75)
    min_val = group.min()
    max_val = group.max()
    return pd.Series({
        'count': count, 'mean': mean, 'std': std,
        'ci_high': ci_high, 'ci_low': ci_low,
        'min': min_val, '25%': q1, 'median': median,
        '75%': q3, 'max': max_val
    })

# 🔹 4. 그룹화 및 통계 계산
grouped = df.groupby(['용도지역', '지목'])['단가'].apply(stats).unstack()
zones = df.groupby('용도지역')['단가'].count().sort_values(ascending=False).index.tolist()
uses = df.groupby('지목')['단가'].count().sort_values(ascending=False).index.tolist()

# 🔹 5. 엑셀 열고 분석결과 덮어쓰기
wb = load_workbook(save_path)
ws = wb['분석표']

# 🔹 헤더 작성
for j, use in enumerate(uses):
    ws.cell(row=1, column=2 + j * 2, value=use)

# 🔹 본문 작성
current_row = 2
for zone in zones:
    zone_written = False
    for j, use in enumerate(uses):
        if (zone, use) in grouped.index:
            row_data = grouped.loc[(zone, use)].round(2)
            left = ['count', 'mean', 'std', 'ci_high', 'ci_low']
            right = ['min', '25%', 'median', '75%', 'max']

            for k in range(5):
                if not zone_written and k == 0:
                    ws.cell(row=current_row + k, column=1, value=zone)

                ws.cell(row=current_row + k, column=2 + j * 2, value=row_data[left[k]])
                ws.cell(row=current_row + k, column=2 + j * 2 + 1, value=row_data[right[k]])
            zone_written = True
    current_row += 5

# 🔹 6. 저장
wb.save(save_path)
```
