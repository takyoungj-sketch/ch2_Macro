```python
import os
import glob
import pandas as pd

# -----------------------------
# 기준 폴더
# -----------------------------
base_path = r'C:\startcoding\GUKTO\연립다세대_매매'

# 통합 폴더 생성
merge_folder = os.path.join(base_path, '연립다세대_매매_통합')
os.makedirs(merge_folder, exist_ok=True)

# 시도 폴더 목록
folders = [
    f for f in os.listdir(base_path)
    if os.path.isdir(os.path.join(base_path, f))
    and f != '연립다세대_매매_통합'
]

merged_df = pd.DataFrame()

# -----------------------------
# 파일 순회
# -----------------------------
for region in folders:

    region_path = os.path.join(base_path, region)
    excel_files = glob.glob(os.path.join(region_path, '*.xlsx'))

    for file in excel_files:

        try:
            df = pd.read_excel(
                file,
                skiprows=13,   # 14행부터 데이터
                header=None
            ).iloc[:, :21]     # U열까지

            merged_df = pd.concat([merged_df, df], ignore_index=True)

            print(f"추가 완료: {region} - {os.path.basename(file)}")

        except Exception as e:
            print(f"오류: {file} / {e}")

# -----------------------------
# 저장
# -----------------------------
output_path = os.path.join(
    merge_folder,
    '연립다세대_매매_전국통합.xlsx'
)

merged_df.to_excel(output_path, index=False, header=False)

print("\n전국 통합 완료")
print(output_path)
```

```python
import pandas as pd
import numpy as np
import os

# -----------------------------
# 경로 설정
# -----------------------------
input_file = r"C:\startcoding\GUKTO\연립다세대_매매\연립다세대_매매_통합\연립다세대_매매_전국통합.xlsx"
output_folder = r"C:\startcoding\GUKTO\연립다세대_매매\연립다세대_매매_정제"
os.makedirs(output_folder, exist_ok=True)
output_file = os.path.join(output_folder, "연립다세대_매매_정제.xlsx")

# -----------------------------
# 데이터 읽기
# -----------------------------
df = pd.read_excel(input_file, header=None)

# 안전하게 데이터를 가져오기 위한 보조 함수
def get_col(target_df, idx, default=None):
    return target_df.iloc[:, idx] if target_df.shape[1] > idx else default

df_clean = pd.DataFrame()

# -----------------------------
# 컬럼 매핑 (원본 U열 구조 기준)
# -----------------------------
df_clean['시군구'] = get_col(df, 1)    # B: 시군구
df_clean['번지'] = get_col(df, 2)      # C: 번지
df_clean['건물명'] = get_col(df, 5)    # F: 건물명
df_clean['전용면적'] = get_col(df, 6)  # G: 전용면적 (연립은 G열이 전용)
df_clean['계약년월'] = get_col(df, 8)  # I: 계약년월
df_clean['계약일'] = get_col(df, 9)    # J: 계약일
df_clean['거래금액'] = get_col(df, 10) # K: 거래금액
df_clean['층'] = get_col(df, 11)       # L: 층
df_clean['건축년도'] = get_col(df, 14) # O: 건축년도
df_clean['도로명'] = get_col(df, 15)   # P: 도로명
df_clean['주택유형'] = get_col(df, 20) # U: 주택유형 (인덱스 20)

# -----------------------------
# 1. 해제 거래 제거 (Q열: 인덱스 16)
# -----------------------------
cancel_idx = 16 
if df.shape[1] > cancel_idx:
    # 날짜(8자리)가 적혀있는 행은 해제된 거래이므로 제거
    cancel_val = df.iloc[:, cancel_idx].astype(str).str.strip()
    df_clean = df_clean[~cancel_val.str.match(r'^\d{8}$')]

# -----------------------------
# 2. 숫자 변환 및 데이터 클리닝
# -----------------------------
df_clean['거래금액'] = df_clean['거래금액'].astype(str).str.replace(',', '')
num_cols = ['전용면적','계약년월','계약일','거래금액','층','건축년도']

for col in num_cols:
    df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce')

# -----------------------------
# 3. 날짜 및 연식 계산
# -----------------------------
# 계약년월(202112) + 계약일(31) 조합 -> 20211231
temp_date_str = df_clean['계약년월'].astype(str) + df_clean['계약일'].fillna(1).astype(int).astype(str).str.zfill(2)
temp_dt = pd.to_datetime(temp_date_str, format='%Y%m%d', errors='coerce')

# 출력용 6자리 날짜 (211231)
df_clean['계약일자'] = temp_dt.dt.strftime('%y%m%d')
# 연식 계산
df_clean['연식'] = temp_dt.dt.year - df_clean['건축년도']

# -----------------------------
# 4. 주소 분해 (최대 5개)
# -----------------------------
addr = df_clean['시군구'].astype(str).str.split(' ', expand=True)
for i in range(5):
    df_clean[f'주소{i+1}'] = addr[i] if i < addr.shape[1] else ""

# -----------------------------
# 5. 규모 및 단가 계산
# -----------------------------
df_clean['면적규모'] = (df_clean['전용면적'] / 30).round() * 30
df_clean['연식규모'] = (df_clean['연식'] / 10).round() * 10
# 단가 정수형 변환
df_clean['단가'] = (df_clean['거래금액'] / df_clean['전용면적']).replace([np.inf, -np.inf], np.nan).fillna(0).astype(int)

# -----------------------------
# 6. 최종 컬럼 정리 및 저장
# -----------------------------
df_clean = df_clean.reset_index(drop=True)
df_clean['순서번호'] = df_clean.index + 1

final_cols = [
    '순서번호', '주소1', '주소2', '주소3', '주소4', '주소5',
    '번지', '도로명', '주택유형', '건물명', '계약일자',
    '거래금액', '면적규모', '연식규모', '전용면적', '연식', '층', '단가'
]

# 결과 파일 저장
df_final = df_clean[final_cols]
df_final.to_excel(output_file, index=False)

print(f"정제 완료! 결과 파일: {output_file}")
```
